```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load image and convert to grayscale
im = Image.open('image.png').convert('L')
```

```python
# Convert image to numpy array
im_matrix = np.array(im)
```

```python
# Perform SVD
U, S, VT = np.linalg.svd(im_matrix)
```

```python
# Set k, the number of singular values to keep
k = 50
```

```python
# Truncate the matrices
U_k = U[:, :k]
S_k = np.diag(S[:k])
VT_k = VT[:k, :]
```

```python
# Reconstruct the image
im_recon = np.dot(U_k, np.dot(S_k, VT_k))
```

```python
# Plot the original and reconstructed images
plt.figure(figsize=(20,20))
plt.subplot(1, 2, 1)
plt.imshow(im_matrix, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(im_recon, cmap='gray')
plt.title('Reconstructed Image with k = {}'.format(k))
plt.show()
```

We can adjust the value of k to change the level of compression. A smaller value of k will result in more compression and a lower quality reconstructed image.
